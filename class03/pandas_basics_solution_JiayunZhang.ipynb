```python
# load two data files
general_tb = pd.read_csv('intro_to_data_science_2017/data/students/file1.csv')
study_tb = pd.read_csv('intro_to_data_science_2017/data/students/file2.csv')

general_tb
```

```python
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.sort_values.html4

general_tb.sort_values(['age'], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
```

```python
student_merge = general_tb.merge(study_tb, how='inner', on=None, left_on=None, right_on=None, 
                left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False)

# TIPS:
# specify the join column name, make it emplicit insteading of implicit
# make sure others can understand your code
```

```python
student_merge
```

```python
# counts how many students of each gender
counts=student_merge.groupby('gender').size()
counts
```

```python
# to make a copy of the merged df called student_merge_new
student_merge_new = student_merge

# add a new column called 'language_counts' at the end of df
student_merge_new['language_counts'] = student_merge_new['computer_language'].apply(lambda x: x.count(",") + 1)

# anothter method to count
#student_merge_new['language_counts'] = student_merge_new['computer_language'].apply(lambda x: len(x.split(',')))
```
